```python
import pandas as pd
import matplotlib.pyplot as plt
data = pd.read_csv('sales.csv')
```

## 1 Create a line plot for the sales of Product_A over the week.

```python

plt.plot(data["Day"],data['Product_A'])
plt.xlabel("Day of week")
plt.ylabel("sales")
plt.title("sales of product_A over a week")
plt.show()
```

## 2 Create a line plot showing the sales trends for Product_A and Product_B on the same graph.


```python
plt.plot(data["Day"],data["Product_A"],label = "Product_B")
plt.plot(data["Day"],data["Product_B"],label ="Product_A")
plt.xlabel("Day of Week")
plt.ylabel("sales")
plt.title("Sales trend of Product_A and Product_B")
plt.legend()
plt.show()
```

## 3 Create a bar plot for the sales of Product_A over the week.

```python
plt.bar(data["Day"],data["Product_A"],label = "Product_A", color = "green")
plt.xlabel("Day of Week")
plt.ylabel("Sales")
plt.title("sales of Product_A over the week")
plt.legend()
plt.show()
```

## 4 Create a horizontal bar plot for the sales of Product_B over the week.


```python
plt.barh(data["Day"],data["Product_B"])
plt.xlabel("Sales")
plt.ylabel ("Day of the Week")
plt.title("'Horizontal Bar Plot of Product B Sales")
plt.show()
```

## 5 Create a scatter plot for Product_A sales against Product_B sales.

```python
plt.scatter(data["Product_A"],data["Product_B"])
plt.xlabel("Product_A sales")
plt.ylabel("Product_B sales")
plt.title("Scatter Plot of Product A vs. Product B Sales")
plt.show()
```

## 6 Create a pie chart showing the total sales distribution of Product_A, Product_B, and Product_C over the week.

```python
total_sales = data[["Product_A","Product_B","Product_C"]].sum()
plt.pie(total_sales,labels = total_sales.index,autopct="%1.1f%%")
plt.title("Total Sales Distribution of Products")
plt.show()
```

## 7 Plot the sales of Product_A and Product_B with different line styles (e.g., solid, dashed).

```python
plt.plot(data["Day"],data["Product_A"],linestyle ="--",label = "product_A")
plt.plot(data["Day"],data["Product_B"],linestyle = "-",label = "product_B")
plt.xlabel("Day of week")
plt.ylabel("Sales")
plt.title("Sales of Product_A and Product_B over a week")
plt.legend()
plt.show()


```

## 8 Create an area plot to show the cumulative sales of Product_A, Product_B, and Product_C over the week.


```python
data.plot.area(x="Day",y = ["Product_A","Product_B","Product_C"])
plt.xlabel('Day of the Week')
plt.ylabel('Sales')
plt.title('Area Plot of Sales')
plt.show()
```

## 9 Create two subplots in a single figure: one for Product_A sales and one for Product_B sales.

```python

fig, (plt1,plt2)= plt.subplots(2,1, figsize= (7,7))
plt1.plot(data["Day"],data["Product_A"],label = "Product_A")
plt1.set_title("Sales of Product_A")
plt1.set_xlabel("Day of a Week")
plt1.set_ylabel("Sales")
plt2.plot(data["Day"],data["Product_B"],label = "Product_B")
plt2.set_title("Sales of Product_B")
plt2.set_xlabel("Day of Week")
plt2.set_ylabel("Sales")
plt.tight_layout()
plt.show()
```

## 10 Create three subplots in a single figure: one for Product_A sales , one for Product_B sales and one for Product_C.

```python
fig, (plt1,plt2, plt3)= plt.subplots(3,1, figsize= (7,7))
plt1.bar(data["Day"],data["Product_A"],label = "Product_A", color = "green")
plt1.set_xlabel("Day of Week")
plt1.set_ylabel("Sales")
plt1.set_title("sales of Product_A over the week")
plt1.legend()

plt2.bar(data["Day"],data["Product_B"],label = "Product_B", color = "blue")
plt2.set_xlabel("Day of Week")
plt2.set_ylabel("Sales")
plt2.set_title("sales of Product_B over the week")
plt2.legend()


plt3.bar(data["Day"],data["Product_C"],label = "Product_C", color = "red")
plt3.set_xlabel("Day of Week")
plt3.set_ylabel("Sales")
plt3.set_title("sales of Product_C over the week")
plt3.legend()


plt.tight_layout()
plt.subplots_adjust(hspace=0.5)
plt.show()

```
